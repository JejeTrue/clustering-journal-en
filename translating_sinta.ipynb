```python
# prompt: saya memiliki 12 data yaitu data_1 - data_12, saya ingin menggabungkan ke 12 dataset tersebut dalam 1 dataset

import pandas as pd
import os

# Define the directory containing the CSV files
directory = '/content/drive/MyDrive/skripsi/full_en/data_translate'  # Replace with your actual directory

# Create an empty list to store the dataframes
dfs = []

# Loop through the files in the directory
for i in range(1, 13):  # Assuming you have files named data_1.csv to data_12.csv
    filename = f'data_{i}.csv'
    filepath = os.path.join(directory, filename)

    # Check if the file exists before attempting to read it
    if os.path.exists(filepath):
        try:
            df = pd.read_csv(filepath)
            dfs.append(df)
        except pd.errors.EmptyDataError:
            print(f"Warning: File '{filename}' is empty. Skipping.")
        except pd.errors.ParserError:
            print(f"Warning: Could not parse '{filename}'. Skipping.")
    else:
        print(f"Warning: File '{filename}' not found. Skipping.")

# Concatenate all the dataframes into a single dataframe
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)

    # Save the combined dataframe to a new CSV file
    output_path = os.path.join(directory, 'combined_data.csv')
    combined_df.to_csv(output_path, index=False)
    print(f"Combined data saved to: {output_path}")
else:
    print("No valid data files found in the specified directory.")
```
